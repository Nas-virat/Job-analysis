Create column extract skill from job description

```python
import re
# Define the language skills
SKILLS = [
    'python',
    'javascript'
    'java',
    'react',
    'react.js',
    'next.js',
    'angular',
    'node.js',
    '.net',
    'c#',
    'sql',
    'gitlab',
    'golang',
    'ci/cd',
    'github action',
    'grafana',
    'vue.js',
    'aws',
    'azure',
    'kafka',
    'microservice',
    'gcp',
]

df['description'] = df['description'].str.lower()

def extract_skill(word):
    result = []
    for skill in SKILLS:
        if skill in word:
            result.append(skill)
    return result

# Extract language skills from job descriptions
df['skills'] = df['description'].apply(extract_skill)

# Display the updated dataframe
df
```
